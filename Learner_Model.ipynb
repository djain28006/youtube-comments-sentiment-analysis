DATA PREPROCESSING

```python
label_names = dataset["train"].features["labels"].feature.names

label_names
```

```python
def decode_labels(example):
    example["emotions"]=[label_names[i] for i in example["labels"]]
    return example
```

```python
train_ds = dataset["train"].map(decode_labels)
val_ds   = dataset["validation"].map(decode_labels)
test_ds  = dataset["test"].map(decode_labels)
```

```python
train_ds[0]
```

```python
train_ds = train_ds.remove_columns(["id", "labels"])
val_ds   = val_ds.remove_columns(["id", "labels"])
test_ds  = test_ds.remove_columns(["id", "labels"])
```

```python
print(train_ds[0])
print(test_ds[0])
print(val_ds[0])
```

```python
from datasets import Dataset

train_ds = train_ds.map(lambda x: {'text': x['text'].lower()})
test_ds = test_ds.map(lambda x: {'text': x['text'].lower()})
val_ds = val_ds.map(lambda x: {'text': x['text'].lower()})
```

```python
train_ds[0]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=20000, oov_token="<OOV>")

tokenizer.fit_on_texts(train_ds['text'])
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.word_index
```

```python
def tokenize_example(example):
    example['tokens'] = tokenizer.texts_to_sequences([example['text']])[0]
    return example

train_ds = train_ds.map(tokenize_example)
val_ds   = val_ds.map(tokenize_example)
test_ds  = test_ds.map(tokenize_example)
```

```python
X_train=pad_sequences(
    train_ds['tokens'],
    maxlen=max_len,
    padding='post'
)

X_val=pad_sequences(
    val_ds['tokens'],
    maxlen=max_len,
    padding='post'
)

X_test=pad_sequences(
    test_ds['tokens'],
    maxlen=max_len,
    padding='post'
)
```

```python
X_train.shape
```

```python
train_ds.features
```

```python
all_emotions = set()
for emotions in train_ds['emotions']:
    for e in emotions:
        all_emotions.add(e)

all_emotions = sorted(list(all_emotions))

num_labels = len(all_emotions)   

emotion2idx = {emotion: idx for idx, emotion in enumerate(all_emotions)}
idx2emotion = {idx: emotion for emotion, idx in emotion2idx.items()}
```

```python
import numpy as np

def encode_multi_hot(example):
    label_vector = np.zeros(num_labels, dtype=np.float32)
    for emotion in example['emotions']:
        label_vector[emotion2idx[emotion]] = 1.0
    return {'label_vector': label_vector}
```

```python
train_ds = train_ds.map(encode_multi_hot)
val_ds   = val_ds.map(encode_multi_hot)
test_ds  = test_ds.map(encode_multi_hot)
```

```python
train_ds[40]
```

```python
y_train = np.array(train_ds['label_vector'])
y_val   = np.array(val_ds['label_vector'])
y_test  = np.array(test_ds['label_vector'])
```

MODEL TRAINING

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional, Input, SpatialDropout1D
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

vocab_size = 28082
embed_dim = 100
max_len = 27

model = Sequential()
model.add(Input(shape=(max_len,)))

# 1. Embedding Layer
model.add(Embedding(vocab_size, embed_dim))

# 2. SpatialDropout (Better for NLP than standard Dropout)
# Drops entire 1D feature maps instead of individual elements
model.add(SpatialDropout1D(0.3))

# 3. First LSTM Layer
# Reduced units from 128 to 64 to prevent overfitting
model.add(
    Bidirectional(
        LSTM(64, return_sequences=True) 
    )
)

# 4. Second LSTM Layer
# Reduced units from 64 to 32
model.add(
    Bidirectional(
        LSTM(32)
    )
)

# 5. Stronger Dropout
model.add(Dropout(0.5))

# 6. Dense Layer with L2 Regularization
# kernel_regularizer=l2(0.001) penalizes large weights
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.001)))

# 7. Output Dropout
model.add(Dropout(0.5))

# 8. Output Layer
model.add(Dense(num_labels, activation='sigmoid'))

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
# EarlyStopping: Stops if no improvement for 5 epochs
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# --- Training ---
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=20,          # Increased slightly since we have EarlyStopping
    batch_size=64,
    callbacks=[early_stopping]
)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
threshold = 0.5

def decode_prediction(prob_vector):
    emotions = []
    for idx, prob in enumerate(prob_vector):
        if prob >= threshold:
            emotions.append(idx2emotion[idx])
    return emotions
```

```python
sample_id = 22

probs = y_pred[sample_id]
print(probs)
predicted_emotions = decode_prediction(probs)

print(predicted_emotions)
```

```python
def decode_text(token_ids):
    words = []
    for token in token_ids:
        if token != 0:  
            words.append(tokenizer.index_word.get(token, "<UNK>"))
    return " ".join(words)
```

```python
i = 12
def decode_prediction_topk(prob_vector, k=2):
    top_indices = prob_vector.argsort()[-k:][::-1]
    return [idx2emotion[i] for i in top_indices]

# Original text (decoded from tokens)
text = decode_text(X_test[i])

# Model prediction
probs = y_pred[i]
emotions = decode_prediction_topk(probs)

print("Comment:")
print(text)

print("\nPredicted emotions:")
print(emotions)
```
